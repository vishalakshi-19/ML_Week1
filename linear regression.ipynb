LINEAR REGRESSION 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv("C:/Users/visha/Downloads/archive (1).zip")
```

```python
x = data[['hours_studied']]
y = data['exam_score']
model=LinearRegression()
model.fit(x,y)
hours=int(input("enter the number hours studied"))
predicted_scores= model.predict([[hours]])[0]

print(f"for {hours} hours studied, the predicted exam score is:{predicted_scores:.2f}")
```

```python
plt.figure(figsize=(10,6))
plt.hist(data["exam_score"],bins=30,color='skyblue',edgecolor='black')
plt.xlabel("Exam scores")
plt.ylabel("number of students")
plt.grid(True)
plt.show()
```
