```python
import pandas as pd
```

```python
import numpy as np
```

```python
 data = {
 'Name': ['Alice', 'Bob', 'Carol', 'Dave', 'Eva', 'Frank', 'Grace'],
 'Age': [23, 25, 22, 24, 21, 26, 23],
 'Gender': ['F', 'M', 'F', 'M', 'F', 'M', 'F'],
 'Marks': [85, 58, 90, 45, 76, 67, 92],
 'Department': ['CSE', 'ECE', 'CSE', 'ME', 'ECE', 'CSE', 'CSE']
 }
```

```python
df=pd.DataFrame(data)
df
```

```python
df.shape
```

```python
df.info
```

```python
df.describe()
```

```python
df["Result"]=np.where(df['Marks']>=60,'pass','fail')
df
```

```python
def assign_grade(marks):
    if marks >= 90:
        return 'A'
    elif marks >= 80:
        return 'B'
    elif marks >= 70:
        return 'C'
    elif marks >= 60:
        return 'D'
    else:
        return 'F'

df['Grade'] = df['Marks'].apply(assign_grade)
print(df[['Name', 'Marks', 'Grade']])
```

```python



avg_marks = df.groupby('Department')['Marks'].mean()
print("Average by departments:")
print(avg_marks)


count_students = df['Department'].value_counts()
print("Count of Students:")
print(count_students)
```

```python
fem_cse_high = df[
    (df['Gender'] == 'F') &
    (df['Department'] == 'CSE') &
    (df['Marks'] > 80)
]
print("Female CSE students with Marks > 80:")
print(fem_cse_high)
```

```python
top_grade_students = df[df['Grade'].isin(['A', 'B'])]
print("Students with Grade A or B:")
print(top_grade_students)
```

```python
import matplotlib.pyplot as plt
```

```python

dept_counts = df['Department'].value_counts()
```

```python

plt.figure(figsize=(6, 4))
dept_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Students per Department')
plt.xlabel('Department')
plt.ylabel('Number of Students')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python

result_counts = df['Result'].value_counts()


plt.figure(figsize=(5, 5))
result_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'salmon'])
plt.title('Pass vs Fail Distribution')
plt.ylabel('')  
plt.tight_layout()
plt.show()
```

```python
# Histogram
plt.figure(figsize=(6, 4))
plt.hist(df['Marks'], bins=10, color='orange', edgecolor='black')
plt.title('Distribution of Marks')
plt.xlabel('Marks')
plt.ylabel('Number of Students')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python

grade_counts = df['Grade'].value_counts()
dept_counts = df['Department'].value_counts()


fig, axes = plt.subplots(1, 3, figsize=(18, 5))

axes[0].hist(df['Marks'], bins=10, color='skyblue', edgecolor='black')
axes[0].set_title('Marks Distribution')
axes[0].set_xlabel('Marks')
axes[0].set_ylabel('Frequency')


axes[1].pie(grade_counts, labels=grade_counts.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
axes[1].set_title('Grade Distribution')

axes[2].bar(dept_counts.index, dept_counts.values, color='lightcoral')
axes[2].set_title('Students per Department')
axes[2].set_xlabel('Department')
axes[2].set_ylabel('Count')

plt.tight_layout()
plt.show()
```
