```python

import matplotlib.pyplot as plt



iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
```

```python
plt.figure(figsize=(6, 4))
plt.hist(df['sepal length (cm)'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
species_counts = df['species'].value_counts()
plt.bar(species_counts.index, species_counts.values, color='lightcoral', edgecolor='black')
plt.title('Species Count')
plt.xlabel('Species')
plt.ylabel('Count')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
colors = {'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}

for species in df['species'].unique():
    subset = df[df['species'] == species]
    plt.scatter(subset['sepal length (cm)'], subset['petal length (cm)'],
                label=species, color=colors[species])

plt.title('Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend(title='Species')
plt.grid(True)
plt.tight_layout()
plt.show()
```
